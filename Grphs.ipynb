```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

lecx=df1['left_eye_center_x']
lecy=df1['left_eye_center_y']
 
info = pd.DataFrame({ 
'Left Eye x': lecx,  
'Left Eye y': lecy
})  
ax = info.hist(bins=4)
```

```python
print("mean left x ",df1['left_eye_center_x'].mean())
print("median left x ",df1['left_eye_center_x'].median())
print("mode left x\n",df1['left_eye_center_x'].mode())



print("mean left y ",df1['left_eye_center_y'].mean())
print("median left y ",df1['left_eye_center_y'].median())
print("mode left y \n",df1['left_eye_center_y'].mode())
```

```python
df = pd.read_csv('roadAccStats13-16.csv')
plt.figure(figsize = (20,10))
y = df['State/UT-Wise Total Number of Road Accidents during - 2016']
yd = df['States/UTs']
p =df['States/UTs'].nunique()
d = np.linspace(1,p,p)   
plt.bar(d, y, align = 'center')
plt.xticks(d, yd, rotation = 90)
plt.xlabel('States')
plt.ylabel('Number of Accidents')
plt.title('Total number of Accidents in each State.')
plt.show()
```
